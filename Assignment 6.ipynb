```python
import pandas as pd
from plotnine import *

collisions= pd.read_csv ("https://raw.githubusercontent.com/mandiemannz/Data-608/master/NYPD_Motor_Vehicle_Collisions%202017.csv")
```

```python
collisions.head(20)
```

Here you will see the collisions dataset. It appears as if Passenger Vechicle has the most collisions followed by SUV/Station Wagons and Pick-up Trucks. 

```python
collisions['VEHICLE TYPE CODE 1'].value_counts().sort_index().plot.bar()
```

The Zip Code with most collisions is 10306 and the least amount of collisions is at 10302.

```python
collisions['ZIP CODE'].value_counts().sort_index().plot.bar()
```

According to this dataset, The most collisions occurred on Staten Island (reason being, this dataset is only Staten Island)

```python
collisions['BOROUGH'].value_counts().sort_index().plot.bar()
```

Finally, the most collisions took place from -74.2 - -74.1 Longitude and Latitude, followed by a few outliers between -74.0 - -73.9

```python
import matplotlib.pyplot as plt
plt.scatter(x=collisions['LONGITUDE'], y=collisions['LATITUDE'])
plt.show()
```
